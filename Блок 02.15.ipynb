## Задание номер 1
##### По файлу посчитайте кол-во победителей финальных олимпиад (stage = 4) по математике

```python
import pandas as pd
df = pd.read_csv('Files/moscow schools - winners.csv', encoding='utf-8', delimiter=',')
df
```

```python
df[(df.Stage == '4') & (df.Status == 'победитель') & (df.Subject == 'Математика')].FullName.count()
```

## Задание номер 2
##### По тому же файлу посчитайте кол-во призеров регионального этапа (stage=3) по информатике из 9-ого класса

```python
df[(df.Stage == '3') & (df.Status == 'призёр') & (df.Subject == 'Информатика') & (df.Class == '9')].FullName.count()
```
